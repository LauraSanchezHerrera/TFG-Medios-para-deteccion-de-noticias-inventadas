Este programa se va a encargar de construir el **dendograma final de las dependencias que se encuentren en relación con una noticia**. Tras construir la **"Base de datos"** con las **dependencias relativas a cada noticia**, comparando una a una todas las noticias que se han recopilado, se van a **agrupar las noticias según las dependencias encontradas**.

**¿Cómo se va a hacer?**

- Dada una noticia se va a sacar, gracias a la matrices de distancia generadas, las noticias con las que tiene:
 - Al menos una de las dos distancias (n1 --> n2 NCD(n1,n2) o n2 --> n1 NCD(n2, n1) debe ser menor que 0,765.
 - Y la otra distancia debe ser menor que 0,8.
- Cuando se tiene este primer listado de noticias con las que tiene similitud, es decir, tienen que ver sus contenidos, se hará lo mismo para todas esas noticias. Obteniendo una **lista de noticias que tienen relación con la noticia inicial**.
- Finalmente, se generará la **matriz de distancias entre las noticias de la lista** y formaré el **dendograma con los diferentes clusters**.

Resultado final del programa: **DENDOGRAMA DE NOTICIAS QUE ESTÁN RELACIONADAS**

Este fichero contiene varios métodos, uno de los más importantes es el método **cluster_una_noticia** encargado de generar un dendograma relacionando en clusters el conjunto de noticias similares obtenidas a partir de la noticia que se pasa por argumento:

```python
# Genera el dendograma resultante del conjunto de noticias relacionadas con la noticia que se pasa por argumento
# Recibe el periodico donde esta la noticia y el numero con la que se ha guardado.
# También la lista de noticias que ya forman parte de un dendograma
def cluster_una_noticia(periodico, noticia, lista_noticias):
    periodico_buscado = periodico # Periodico donde esta la noticia de la que se quiere generar el dendograma
    noticia_buscada = noticia # Numero de la noticia de la que se quiere generar el dendograma
    
    # Creo el fichero donde voy a ir escribiendo el procedimiento del estudio de esta noticia
    f = open('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'w')
    f.write('Estudio de la noticia ' + periodico_buscado + "-" + noticia_buscada + '\n\n')
    
    # Librerias utilizadas
    import os
    from datetime import datetime
    
    # Primero extraigo las diferentes carpetas que se hayan generado

    # Listado de las carpetas que hay en el directorio "Base de datos"
    carpetas_dependencias = os.listdir('Noticias/Base de datos')

    # Extraigo el nombre de todos los directorios (noticias) que hay en cada carpeta de dependencia
    dicc_carpetas_dependencias_noticias = {} # Para cada carpeta de dependencias guarda la lista de noticias

    for carpeta_dep in carpetas_dependencias:
        dicc_carpetas_dependencias_noticias[carpeta_dep] = os.listdir('Noticias/Base de datos/' + carpeta_dep)
    
    # Recorro los directorios en busqueda de la noticia
    nombre_noticia_buscada = periodico_buscado + "-" + noticia_buscada # nombre que tiene el directorio

    # Cabe la posibilidad de que la noticia no tenga datos en la base de datos, porque sea demasiado corta
    # o demasiado larga en comparacion con el resto. Fue imposible compararla con otras.

    # Recorro todos los directorios hasta encontrar en que carpeta de dependencias esta
    carpeta_dependencia_pertenece = "No encontrado"
    for carpeta_dep in dicc_carpetas_dependencias_noticias:
        for dir_noticia in dicc_carpetas_dependencias_noticias[carpeta_dep]:
            if dir_noticia == nombre_noticia_buscada:
                carpeta_dependencia_pertenece = carpeta_dep

    # Si no se ha encontrado acaba aqui
    if carpeta_dependencia_pertenece == "No encontrado":
        lista_noticias.append(nombre_noticia_buscada)
        f.close()
        os.remove('Noticias/diciembre 2021/día 4/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt')
        return

    # Ya puedo sacar las matrices y las posiciones en la matriz
    list_matrices = [] # Estructura para guardar las matrices
    list_posiciones = [] # Estructura para guardar las posiciones

    list_informacion_noticia = os.listdir('Noticias/Base de datos/' + carpeta_dependencia_pertenece + '/' + nombre_noticia_buscada)

    # Compruebo que sea correcto
    f.write("Pertenece a la carpeta de dependencias: " + carpeta_dependencia_pertenece + '\n\n')

    # Librerias utilizadas
    import json
    import numpy as np
    
    # Extraigo las matrices y las posiciones en la matriz, el cluster solo me sirve para verlo graficamente
    for documento in list_informacion_noticia:
        # url donde encontrar el documento
        url = 'Noticias/Base de datos/' + carpeta_dependencia_pertenece + '/' + nombre_noticia_buscada + '/' + documento
        if documento.split(" ")[0] == "Matriz":
            with open(url) as file:
                list_matrices.append(np.array(json.load(file))) # Vuelvo a transformarlo a una matriz
        if documento.split(" ")[0] == "Posicion":
            with open(url) as file:
                list_posiciones.append(json.load(file))

    # Librerias utilizadas
    import json
    import os
    
    # Guardo las noticias recopiladas de cada periodico en una variable diferente
    # EL MUNDO
    noticias_periodico_ELMUNDO = {}
    url_noticias_periodico_ELMUNDO = 'Noticias/EL MUNDO/noticias.json'
    with open(url_noticias_periodico_ELMUNDO) as file:
        noticias_periodico_ELMUNDO = json.load(file)

    # 20 minutos
    noticias_periodico_20minutos = {}
    url_noticias_periodico_20minutos = 'Noticias/20 minutos/noticias.json'
    with open(url_noticias_periodico_20minutos) as file:
        noticias_periodico_20minutos = json.load(file)

    # elEconomista.es
    noticias_periodico_elEconomistaes = {}
    url_noticias_periodico_elEconomistaes = 'Noticias/elEconomista.es/noticias.json'
    with open(url_noticias_periodico_elEconomistaes) as file:
        noticias_periodico_elEconomistaes = json.load(file)

    # La Informacion
    noticias_periodico_LaInformacion = {}
    url_noticias_periodico_LaInformacion = 'Noticias/La Informacion/noticias.json'
    with open(url_noticias_periodico_LaInformacion) as file:
        noticias_periodico_LaInformacion = json.load(file)

    # El Confidencial
    noticias_periodico_ElConfidencial = {}
    url_noticias_periodico_ElConfidencial = 'Noticias/El Confidencial/noticias.json'
    with open(url_noticias_periodico_ElConfidencial) as file:
        noticias_periodico_ElConfidencial = json.load(file)


    dicc_periodico_noticias = {} # Diccionario para almacenar las noticias de cada periodico
    dicc_periodico_noticias['EL MUNDO'] = noticias_periodico_ELMUNDO
    dicc_periodico_noticias['20 minutos'] = noticias_periodico_20minutos
    dicc_periodico_noticias['elEconomista.es'] = noticias_periodico_elEconomistaes
    dicc_periodico_noticias['La Informacion'] = noticias_periodico_LaInformacion
    dicc_periodico_noticias['El Confidencial'] = noticias_periodico_ElConfidencial
    
    # Recorro cada matriz buscando las noticias que cumplan las condiciones definidas para que sea dependiente de la noticia buscada
    # Tengo que recorrer la primera fila y columna pero sin tener en cuenta la primera posicion
    
    periodicos = set() # Conjunto de los periodicos en los que se han encontrado las dependencias
    periodicos.add(periodico_buscado)

    noticias_parecidas = set() # Conjunto de noticias similares a la noticia buscada
    mismo_articulo = [] # Lista de noticias que son el mismo articulo, con tener en cuenta 1 es suficiente
    num_matriz = 0 # Indica el numero de matriz que se este mirando
    for matriz in list_matrices:
        # Recorro la primera fila
        # Compruebo que la noticia cumpla las condiciones de dependencia
        parecida = [] # Tendra el valor 2 si cumple que en al menos un caso la distancia < 0,765 y en el otro < 0,8
        for n in range(0, len(matriz)-1):
            parecida.append(0) 

        num_columna = -1
        for distancia in matriz[0]:
            if distancia != 0 and distancia < 0.765:
                parecida[num_columna] += 1
            num_columna += 1

        # Recorro la primera columna
        num_columna = -1
        for fila in matriz:
            if fila[0] != 0 and fila[0] < 0.765:
                parecida[num_columna] += 1
            num_columna += 1

        num_columna = 1
        for valor_dep in parecida:
            if valor_dep == 2:
                noticias_parecidas.add(list_posiciones[num_matriz][num_columna])
            if valor_dep == 1:
                distancia_columna = matriz[0][num_columna]
                distancia_fila = matriz[num_columna][0]
                if distancia_fila < 0.81 and distancia_columna < 0.81:
                    noticias_parecidas.add(list_posiciones[num_matriz][num_columna])
            if (matriz[0][num_columna] + matriz[num_columna][0])/2 < 0.781:
                noticias_parecidas.add(list_posiciones[num_matriz][num_columna])
            num_columna += 1

        num_matriz += 1
    
        
    # Compruebo no se trata del mismo articulo y que las fechas son relativamente del mismo tiempo
    noticias_a_eliminar = []
    for noticia in noticias_parecidas:
        f.write("La noticia es similar a la noticia "  + noticia + " teniendo en cuenta la NCD. Pero hay que realizar las siguientes comprobaciones:" + '\n')
        f.write(" - Comprobando si la noticia " + nombre_noticia_buscada + " es el mismo artículo que la noticia " + noticia + '\n')
        
        periodico1 = noticia.split("-")[0]
        noticia1 = noticia.split("-")[1]

        flag = 0 # Si cumplen ambas condiciones la flag tendra el valor 2

        if dicc_periodico_noticias[periodico1][noticia1]["url"] != dicc_periodico_noticias[periodico_buscado][noticia_buscada]["url"]:
            flag += 1
            f.write("     Se trata de noticias distintas: \n")
            f.write("        - " + dicc_periodico_noticias[periodico1][noticia1]["url"]+ '\n')
            f.write("        - " + dicc_periodico_noticias[periodico_buscado][noticia_buscada]["url"] + '\n\n')
        else:
            mismo_articulo.append(noticia)
            f.write("     Se trata de la misma noticia: \n")
            f.write("        - " + dicc_periodico_noticias[periodico1][noticia1]["url"]+ '\n')
            f.write("        - " + dicc_periodico_noticias[periodico_buscado][noticia_buscada]["url"]+ '\n\n')
        
        f.write(" - Comprobando si las fechas son relativamente del mismo tiempo, máximo 7 días de diferencia" + '\n')
        
        if dicc_periodico_noticias[periodico1][noticia1]["fecha"] != "Muy antigua" and dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"] != "Muy antigua":
            fecha = dicc_periodico_noticias[periodico1][noticia1]["fecha"].split("-")[0] + "/" +dicc_periodico_noticias[periodico1][noticia1]["fecha"].split("-")[1] + "/" +dicc_periodico_noticias[periodico1][noticia1]["fecha"].split("-")[2].split("T")[0]
            fecha_buscada = dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"].split("-")[0] + "/" +dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"].split("-")[1] + "/" +dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"].split("-")[2].split("T")[0]
            fecha_1 = datetime.strptime(fecha, '%Y/%m/%d')
            fecha_2 = datetime.strptime(fecha_buscada, '%Y/%m/%d')

            if str(fecha_1 - fecha_2) == "0:00:00":
                flag += 1
                f.write("     Se trata de la misma fecha: \n")
                f.write("        - " + fecha + '\n')
                f.write("        - " + fecha_buscada + '\n\n')
            else:
                
                if fecha_1 < fecha_2:
                    if int(str(fecha_1 - fecha_2).split(" ")[0].split("-")[1]) < 8:
                        flag += 1
                        
                else:
                    if int(str(fecha_1 - fecha_2).split(" ")[0]) < 8:
                        flag += 1
                        
                
                if fecha_1 < fecha_2:
                    f.write("     Tienen " + str(fecha_1 - fecha_2).split(" ")[0].split("-")[1] + " días de diferencia: \n")
                else:
                    f.write("     Tienen " + str(fecha_1 - fecha_2).split(" ")[0] + " días de diferencia: \n")
                f.write("        - " + fecha + '\n')
                f.write("        - " + fecha_buscada + '\n\n')

        if flag != 2:
            noticias_a_eliminar.append(noticia)
            f.write("Demasiados días de diferencia o se trata del mismo artículo.\n\n")
        else:
            f.write("Cumple todos los aspectos, sí se trata de una noticia similar.\n\n")

    # Elimino las noticias que no cumplan las condiciones
    for noticia in noticias_a_eliminar:
        noticias_parecidas.remove(noticia)
    
    f.write("Se han encontrado las siguientes noticias parecidas:\n")
    f.write(str(noticias_parecidas) + "\n")
    f.write("Pero en el caso de haber varias es posible que entre ellas sean el mismo artículo.\n\n")
    f.write("Comprobando si es el caso...\n")
            
    
    # Compruebo si las dependencias encontradas son la misma noticia
    pares_de_noticias_iguales = {}
    for noticia in noticias_parecidas:

        periodico1 = noticia.split("-")[0]
        noticia1 = noticia.split("-")[1]

        for noticia_1 in noticias_parecidas:
            periodico2 = noticia_1.split("-")[0]
            noticia2 = noticia_1.split("-")[1]

            if noticia1 != noticia2:
                
                f.write(" - Comparo las noticias " + noticia + " y " + noticia_1 + ":\n")
                
                if dicc_periodico_noticias[periodico1][noticia1]["url"] == dicc_periodico_noticias[periodico2][noticia2]["url"]:
                    f.write("     Son la misma noticia, por lo tanto solo tendré en cuenta una de ellas.\n\n")
                    if noticia not in pares_de_noticias_iguales:
                        pares_de_noticias_iguales[noticia] = [noticia_1]
                    else:
                        pares_de_noticias_iguales[noticia].append(noticia_1)
                else:
                    f.write("     No son la misma noticia.\n\n")

    par_eliminado = set()
    noticia_eliminada = set()
    for noticia in pares_de_noticias_iguales:
        if noticia not in noticia_eliminada:
            for noticia_1 in pares_de_noticias_iguales[noticia]:
                if noticia_1 not in par_eliminado:
                    par_eliminado.add(noticia)
                    noticia_eliminada.add(noticia_1)
                    noticias_parecidas.remove(noticia_1)
                    mismo_articulo.append(noticia_1)

    f.write("Hay " + str(len(noticias_parecidas)) + " noticias similares:\n")
    f.write(str(noticias_parecidas) + "\n\n")
    
    # No tiene noticias similares
    if len(noticias_parecidas) == 0:
        lista_noticias.append(nombre_noticia_buscada)
        f.write("No tiene noticias similares.\n\n")
        f.close()
        os.remove('Noticias/diciembre 2021/día 4/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt')
        return
    
    # Voy a crear una estructura para ver como van derivando unas noticias a otras
    # Diccionario para almacenar para cada noticia las noticias similares que deriven de ella
    dicc_noticias_similares = {nombre_noticia_buscada: noticias_parecidas}
    f.write("Tras buscar las noticias similares de la noticia buscada, hemos obtenido las siguientes dependencias:\n")
    f.write(str(dicc_noticias_similares) + "\n\n")
    f.write("Bucle recursivo para encontrar las dependencias de cada noticia que salga similar:\n")
    f.close()
    # Lista de noticias parecidas totales
    lista_noticias_parecidas = set()
    for noticia in noticias_parecidas:
        lista_noticias_parecidas.add(noticia)
    lista_noticias_parecidas.add(nombre_noticia_buscada)
    nuevas = True # Indica si se han encontrado noticias similares nuevas
    lista_noticias_estudiar = noticias_parecidas
    while 1 == 1:
        if nuevas == False:
            break

        # Busco nuevas noticias similares de las noticias que todavia no he estudiado
        noticias_nuevas = busca_noticias_similares_matriz(dicc_noticias_similares, lista_noticias_estudiar, dicc_carpetas_dependencias_noticias, dicc_periodico_noticias, mismo_articulo, periodico_buscado, noticia_buscada)
        # noticias_nuevas contiene las noticias similares que han salido a partir de las noticias_parecidas
        # En la siguiente ronda solo hay que estudiar de las noticias_nuevas las que no esten en la lista_noticias_parecidas
        lista_noticias_estudiar = set()
        for noticia_nueva in noticias_nuevas:
            if noticia_nueva not in lista_noticias_parecidas:
                lista_noticias_estudiar.add(noticia_nueva)

        # La union de las noticias_nuevas y la lista lista_noticias_parecidas son el nuevo conjunto de noticias parecidad
        lista_noticias_parecidas = noticias_nuevas | lista_noticias_parecidas

        f = open('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'a')
        f.write("Tras el estudio se han encontrado " + str(len(lista_noticias_estudiar)) + " noticias nuevas similares.\n")

        if len(lista_noticias_estudiar) == 0:
            nuevas = False
            f.write("No hay noticias nuevas que estudiar.\n\n")
            f.close()
        else:
            f.write("Se van a estudiar las noticias nuevas encontradas: \n")
            f.write(str(lista_noticias_estudiar) + '\n\n')
            f.write("Lista de noticias similares actual:\n")
            f.write(str(lista_noticias_parecidas) + '\n\n')
            f.close()


    f = open('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'a')
    f.write("Hay " + str(len(lista_noticias_parecidas)-1) + " noticias que tienen que ver con la noticia buscada:\n")
    f.write(str(lista_noticias_parecidas) + '\n')
    f.write(str(dicc_noticias_similares) + "\n\n")
    f.close()
    
    from ipynb.fs.full.NCD import NCD #importo la funcion que he desarrollado en el fichero NCD.ipynb para calcular la distancia
    from ipynb.fs.full.NCD import write_file_str #importo la funcion que he desarrollado en el fichero NCD.ipynb para escribir el contenido de una noticia a un fichero txt

    filename_noticia_bucle = "noticia_bucle.txt" # nombre del fichero .txt de la noticia_bucle
    filename_noticia_comparada = "noticia_comparada.txt" # nombre del fichero .txt de la noticia_resto
    matriz_NCDs = {} # estructura para almacenar las distancias
    contador_dependencias = 0 # Cuenta el numero de noticias que son parecidas entre si

    # Recorro todas las noticias que se han detectado que son similares
    cont_noticias_comparadas = 0 # Guarda el numero de noticias que se han comparado
    for nombre_noticia_bucle in lista_noticias_parecidas:
        periodico_bucle = nombre_noticia_bucle.split("-")[0]
        num_noticia_bucle = nombre_noticia_bucle.split("-")[1]
        lista_noticias.append(nombre_noticia_bucle)
        periodicos.add(periodico_bucle)

        longitud_bucle = len(dicc_periodico_noticias[periodico_bucle][num_noticia_bucle]['contenido']) # Tamaño noticia buscada

        # Compruebo que la noticia buscada no es demasiado corta
        if longitud_bucle < 1250:
            pass # En este caso no se puede medir la distancia al ser demasiado corta

        write_file_str(filename_noticia_bucle,dicc_periodico_noticias[periodico_bucle][num_noticia_bucle]['contenido']) # Escribo el contenido en un fichero

        # Cada noticia la comparo con el resto de noticias (Si no las he comparado ya)
        cont_aux = 0
        for nombre_noticia_comparada in lista_noticias_parecidas:
            if cont_aux > cont_noticias_comparadas:
                periodico_comparada = nombre_noticia_comparada.split("-")[0]
                num_noticia_comparada = nombre_noticia_comparada.split("-")[1]

                #print("Comparando la noticia " + num_noticia_comparada + " del periodico " + periodico_comparada + " y la noticia " + num_noticia_bucle + " del periodico " + periodico_bucle)

                longitud_comparada = len(dicc_periodico_noticias[periodico_comparada][num_noticia_comparada]['contenido']) # Tamaño noticia comparada

                # Solo comparo la noticia si su contenido no esta vacio y si tiene una longitud mayor de 1250
                if dicc_periodico_noticias[periodico_comparada][num_noticia_comparada]['contenido'] != "" and longitud_comparada >= 1250:
                    # Ademas tiene que tener una longitud parecida con la noticia buscada
                    if longitud_bucle + 1000 >= longitud_comparada and longitud_bucle - 1000 <= longitud_comparada:

                        # Escribo el contenido de la noticia comparada en un fichero txt para compararla con la noticia del bucle
                        write_file_str(filename_noticia_comparada, dicc_periodico_noticias[periodico_comparada][num_noticia_comparada]['contenido']) # Escribo el contenido en un fichero

                        # Calculo las distancias de compresion normalizadas: NCD(noticia_buscada, noticia_comparada) y NCD(noticia_comparada, noticia_buscada)
                        NCDs = NCD(filename_noticia_bucle, filename_noticia_comparada)

                        # Quiero de alguna manera identificar el conjunto de noticias con mas similitud a otras y las que menos
                        # Para ello voy a contar las veces que se cumple la condicion definida anteriormente
                        if (NCDs[0] < 0.765 and NCDs[1] < 0.765) or (NCDs[0] < 0.765 and NCDs[1] < 0.81) or (NCDs[1] < 0.765 and NCDs[0] < 0.81) or ( (NCDs[1] + NCDs[0]) / 2 < 0.781):
                            contador_dependencias += 1

                        # Guardo los resultados en la matriz de distancias

                        # noticia_buscada - noticia_comparada
                        if nombre_noticia_bucle not in matriz_NCDs:
                            matriz_NCDs[nombre_noticia_bucle] = {}
                            matriz_NCDs[nombre_noticia_bucle][nombre_noticia_comparada] = NCDs[0]
                        else:
                            matriz_NCDs[nombre_noticia_bucle][nombre_noticia_comparada] = NCDs[0]

                        # noticia_comparada - noticia_buscada
                        if nombre_noticia_comparada not in matriz_NCDs:
                            matriz_NCDs[nombre_noticia_comparada] = {}
                            matriz_NCDs[nombre_noticia_comparada][nombre_noticia_bucle] = NCDs[1]
                        else:
                            matriz_NCDs[nombre_noticia_comparada][nombre_noticia_bucle] = NCDs[1]

            cont_aux += 1

        cont_noticias_comparadas += 1

    
    contador_dependencias = len(lista_noticias_parecidas) # Numero de noticias similares
    f = open('Noticias/diciembre 2021/día 4/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'a')
    f.write("Paso a calcular la matriz de distancias del conjunto de noticias que he obtenido.\n")
    f.write("Y a crear el cluster a partir de la matriz de distancias calculada.\n\n")
    f.close()
    
    import numpy as np
    import math

    # Calculo el numero de noticias que se han comparado
    numero_noticias_comparadas = len(lista_noticias_parecidas)
    #print("Se han comparado " + str(numero_noticias_comparadas) + " noticias")
    numero_matrices_distancia = numero_noticias_comparadas / 50
    #print("Se van a crear " + str(math.ceil(numero_matrices_distancia)) + " matrices de distancia.")

    # Creo las matrices de distancia que sean necesarias

    list_matrices = [] # lista de matrices
    list_posiciones = [] # lista de posiciones en cada matriz
    contador_noticias_comparadas_matriz = 0 # Indica el numero de noticias comparadas que ya han sido añadidas en matrices

    for numero_matriz in range(0,math.ceil(numero_matrices_distancia)):
        #print("Creando la matriz numero : " + str(numero_matriz))

        matriz_distancias_noticia_buscada_sin_numpy = [] # Estrutura para crear la matriz de distancias

        # Contenido inicial de cada fila de la matriz
        # Primero pongo todo a 1
        # Si el numero de noticias comparadas que quedan son menor que 50, el numero de filas sera menor
        if numero_noticias_comparadas < 50 or numero_noticias_comparadas - contador_noticias_comparadas_matriz < 50:
            if contador_noticias_comparadas_matriz != 0:
                num_filas = numero_noticias_comparadas - contador_noticias_comparadas_matriz
            else:
                num_filas = numero_noticias_comparadas
        else:
            num_filas = 50

        for fila in range(0, num_filas):
            list_aux = []
            for columna in range(0, num_filas):  
                list_aux.append(1)
            matriz_distancias_noticia_buscada_sin_numpy.append(list_aux)

        # Guardo la posicion de cada noticia en la matriz
        aux_contador_noticias_comparadas_matriz = contador_noticias_comparadas_matriz
        limite = contador_noticias_comparadas_matriz + num_filas # Indica hasta que noticia tengo que leer
        cont = 1
        posicion_noticia_matriz = []
        for nombre_noticia in matriz_NCDs:
            if cont > limite:
                break  
            if cont > aux_contador_noticias_comparadas_matriz:
                posicion_noticia_matriz.append(nombre_noticia)
                contador_noticias_comparadas_matriz += 1 # Incremento el numero de noticias en la matriz
            cont += 1

        # Relleno con ceros la diagonal de la matriz ya que una noticia y la misma, son iguales
        cont_fila = 0
        for fila in matriz_distancias_noticia_buscada_sin_numpy:
            fila[cont_fila] = 0
            cont_fila += 1

        # Relleno con las distancias calculadas
        cont_fila = 0
        for nombre_noticia in posicion_noticia_matriz:
            cont_columna = 0
            for nombre_noticia_comparada in posicion_noticia_matriz:
                if nombre_noticia != nombre_noticia_comparada:
                    if nombre_noticia_comparada in matriz_NCDs[nombre_noticia]:
                        matriz_distancias_noticia_buscada_sin_numpy[cont_fila][cont_columna] = matriz_NCDs[nombre_noticia][nombre_noticia_comparada]
                cont_columna += 1
            cont_fila += 1
        
        matriz_distancias_noticia_buscada = np.array(matriz_distancias_noticia_buscada_sin_numpy)

        # Guardo la matriz y las posiciones
        list_matrices.append(matriz_distancias_noticia_buscada)
        list_posiciones.append(posicion_noticia_matriz)
        
    # Gráficos
    # ==============================================================================
    import matplotlib.pyplot as plt
    from matplotlib import style
    style.use('ggplot') or plt.style.use('ggplot')

    # Preprocesado y modelado
    # ==============================================================================
    from sklearn.cluster import AgglomerativeClustering
    from scipy.cluster.hierarchy import dendrogram
    from sklearn.preprocessing import scale
    from sklearn.metrics import silhouette_score

    # Ficheros donde guardar los datos si la noticia no es demasiado corta

    # Carpeta raiz de todos los datos
    if not os.path.exists('Noticias/Base de datos cluster'):
        os.mkdir('Noticias/Base de datos cluster')

    # Carpeta segun el numero de dependencias para diferencias segun el conjunto de noticias con mas dependencias
    if not os.path.exists('Noticias/Base de datos cluster/' + str(contador_dependencias)):
        os.mkdir('Noticias/Base de datos cluster/' + str(contador_dependencias))

    # Carpeta para cada cluster de noticias a raiz de una noticia
    if not os.path.exists('Noticias/Base de datos cluster/' + str(contador_dependencias) + '/' + nombre_noticia_buscada):
        os.mkdir('Noticias/Base de datos cluster/' + str(contador_dependencias) + '/' + nombre_noticia_buscada)
    carpeta_noticia_buscada = 'Noticias/Base de datos cluster/' + str(contador_dependencias) + '/' + nombre_noticia_buscada

    # Funcion para dibujar graficas
    def plot_dendrogram(model, **kwargs):
        '''
        Esta función extrae la información de un modelo AgglomerativeClustering
        y representa su dendograma con la función dendogram de scipy.cluster.hierarchy
        '''

        counts = np.zeros(model.children_.shape[0])
        n_samples = len(model.labels_)
        for i, merge in enumerate(model.children_):
            current_count = 0
            for child_idx in merge:
                if child_idx < n_samples:
                    current_count += 1  # leaf node
                else:
                    current_count += counts[child_idx - n_samples]
            counts[i] = current_count

        linkage_matrix = np.column_stack([model.children_, model.distances_,
                                          counts]).astype(float)

        # Plot
        dendrogram(linkage_matrix, **kwargs)

    # Creo un cluster por cada matriz de distancias
    for numero_matriz in range(0,math.ceil(numero_matrices_distancia)):
        # Modelo utilizado
        # ==============================================================================
        modelo_hclust_complete = AgglomerativeClustering(
                                    affinity = 'precomputed',
                                    linkage  = 'average',
                                    distance_threshold = 0,
                                    n_clusters         = None
                                )
        modelo_hclust_complete.fit(X=list_matrices[numero_matriz])
        
        f = open('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'a')
        f.write("***********Relación número noticia cluster - noticia***********\n\n")
        x = 0
        for noticia in list_posiciones[numero_matriz]:
            f.write(str(x) + ": " + noticia + "\n")
            x += 1
            
        f.write("\n***********    Matriz de distancias calculada    ***********\n\n")
        f.write(str(list_matrices[numero_matriz]) + "\n")
        
        f.write("\n***********Cantidad de información de una noticia a otra***********\n\n")
        
        from ipynb.fs.full.NCD import cantidad_info_n1_en_n2 # funcion para extraer la cantidad de informacion que tiene una noticia de la otra

        filename_1 = "noticia_1.txt" # nombre del fichero .txt de la noticia1
        filename_2 = "noticia_2.txt" # nombre del fichero .txt de la noticia2
        
        x = 0
        for noticia in list_posiciones[numero_matriz]:

            periodico1 = noticia.split("-")[0]
            noticia1 = noticia.split("-")[1]

            for n in dicc_noticias_similares[noticia]:

                periodico2 = n.split("-")[0]
                noticia2 = n.split("-")[1]

                write_file_str(filename_1, dicc_periodico_noticias[periodico1][noticia1]["contenido"])
                write_file_str(filename_2, dicc_periodico_noticias[periodico2][noticia2]["contenido"])

                cant12 = cantidad_info_n1_en_n2(filename_1, filename_2)
                cant21 = cantidad_info_n1_en_n2(filename_2, filename_1)

                f.write("         " + noticia + " -> " + n + ": " + str(matriz_NCDs[noticia][n]) + ". Cantidad de información extraída de la noticia " + noticia + " para añadirla a la noticia " + n + " :" + str(cant12) + "\n")
                f.write("         " + n + " -> " + noticia + ": " + str(matriz_NCDs[n][noticia]) + ". Cantidad de información extraída de la noticia " + noticia + " para añadirla a la noticia " + n + " :" + str(cant21) + "\n\n")
                
            x += 1
        
        f.write("\n***********      Información de las noticias     ***********\n\n")
        f.close()
        for noticia in list_posiciones[numero_matriz]:
            f = open('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'a')
            per = noticia.split("-")[0]
            n = noticia.split("-")[1]
            
            f.write(noticia + " \n")
            
            f.write("Título: \n")
            f.write(dicc_periodico_noticias[per][n]["titulo"]+ "\n")
            f.write("Fecha: \n")
            f.write(dicc_periodico_noticias[per][n]["fecha"]+ "\n")
            f.write("Autor: ")
            if "autor" not in dicc_periodico_noticias[per][n]:
                f.write("No tiene"+ "\n")
            else:
                f.write(dicc_periodico_noticias[per][n]["autor"]+ "\n")
            f.close()
            
            f = open('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'a', encoding="utf-8")
            f.write("Contenido: ")
            f.write(str(dicc_periodico_noticias[per][n]["contenido"]) + "\n\n\n")
            f.close()
        
        f.close()
        

        # Dendrograma
        # ==============================================================================
        fig, figura = plt.subplots(1, 1, figsize=(20,8))
        plot_dendrogram(modelo_hclust_complete, color_threshold=0, ax=figura)
        figura.set_title("Distancia NCD, Linkage average, matriz " + str(numero_matriz))
        plt.tight_layout();

        # Guardo los datos
        # ==============================================================================
        # Solo guardo los datos si la noticia no es demasiado corta
        # Cluster
        nombre_imagen = carpeta_noticia_buscada + "/Cluster " + str(numero_matriz) + ".jpg"
        plt.savefig(nombre_imagen)
        # Matrices de distancia
        # Solo tiene distancias la primera fila y primera columna. Se compara la noticia_buscada con el resto
        nombre_matrices = carpeta_noticia_buscada + "/Matriz " + str(numero_matriz) + ".json"
        with open(nombre_matrices, 'w') as fp:
            json.dump(list_matrices[numero_matriz].tolist(), fp)

        # Lista de posiciones
        # Posicion de la lista = fila en la matriz, la primera fila es de la noticia_buscada
        nombre_posiciones = carpeta_noticia_buscada + "/Posicion " + str(numero_matriz) + ".json"
        with open(nombre_posiciones, 'w') as fp:
            json.dump(list_posiciones[numero_matriz], fp)
            
        # Copio el fichero del estudio al directorio
        import shutil
        shutil.copy('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', carpeta_noticia_buscada + '/Estudio.txt')
        os.remove('Noticias/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt')
        
        # Compruebo si se han encontrado dependencias en mas de un periodico
        if len(periodicos) > 1:
            shutil.copytree(carpeta_noticia_buscada, 'Noticias/Base de datos cluster/Varios periodicos/' + str(contador_dependencias) + '/' + nombre_noticia_buscada)
```

```python
# Funcion que busca noticias similares de un conjunto de noticias
# noticias_parecidas_nuevas -> Conjunto de noticias de las que vamos a sacar a que noticias se parecen 
#                           -> La idea no estudiar dos veces la misma noticia
# dicc_noticias_similares -> Diccionario que guarda para cada noticia, las noticias que se parecen a ella
def busca_noticias_similares_matriz(dicc_noticias_similares, noticias_parecidas_nuevas, dicc_carpetas_dependencias_noticias, dicc_periodico_noticias, mismo_articulo, periodico_buscado, noticia_buscada):
    from datetime import datetime
    
    f = open('Noticias/diciembre 2021/día 4/Base de datos cluster/' + periodico_buscado + "-" + noticia_buscada + '.txt', 'a')

    noticias_parecidas = set()
    #print("Se van a buscar noticias similares de las noticias:")
    #print(noticias_parecidas_nuevas)
    for noticia in noticias_parecidas_nuevas:
        f.write('Estudio de la noticia ' + noticia + '\n\n')
        #print("Buscando la noticia " + noticia)
        carpeta_dependencia_pertenece = "No encontrado"
        for carpeta_dep in dicc_carpetas_dependencias_noticias:
            for dir_noticia in dicc_carpetas_dependencias_noticias[carpeta_dep]:
                if dir_noticia == noticia:
                    carpeta_dependencia_pertenece = carpeta_dep
        
        # Si no se ha encontrado acaba aqui
        if carpeta_dependencia_pertenece == "No encontrado":
            return
        else:
            if noticia not in dicc_noticias_similares:
                dicc_noticias_similares[noticia] = set()
                
        f.write("Pertenece a la carpeta de dependencias: " + carpeta_dependencia_pertenece + '\n\n')
                
        #print("Pertenece a la carpeta de dependencias: " + carpeta_dependencia_pertenece)

        # Ya puedo sacar las matrices y las posiciones en la matriz
        list_matrices = [] # Estructura para guardar las matrices
        list_posiciones = [] # Estructura para guardar las posiciones

        list_informacion_noticia = os.listdir('Noticias/diciembre 2021/día 4/Base de datos más restrictiva/' + carpeta_dependencia_pertenece + '/' + noticia)

        # Extraigo las matrices y las posiciones en la matriz, el cluster solo me sirve para verlo graficamente
        import json
        import numpy as np
        for documento in list_informacion_noticia:
            # url donde encontrar el documento
            url = 'Noticias/diciembre 2021/día 4/Base de datos más restrictiva/' + carpeta_dependencia_pertenece + '/' + noticia + '/' + documento
            if documento.split(" ")[0] == "Matriz":
                with open(url) as file:
                    list_matrices.append(np.array(json.load(file))) # Vuelvo a transformarlo a una matriz
            if documento.split(" ")[0] == "Posicion":
                with open(url) as file:
                    list_posiciones.append(json.load(file))


        num_matriz = 0 # Indica el numero de matriz que se este mirando
        for matriz in list_matrices:
            # Recorro la primera fila
            # Compruebo que la noticia cumpla las condiciones de dependencia
            parecida = [] # Tendra el valor 2 si cumple que en al menos un caso la distancia < 0,765 y en el otro < 0,8
            for n in range(0, len(matriz)-1):
                parecida.append(0) 

            num_columna = -1
            for distancia in matriz[0]:
                if distancia != 0 and distancia < 0.765:
                    parecida[num_columna] += 1
                num_columna += 1

            # Recorro la primera columna
            num_columna = -1
            for fila in matriz:
                if fila[0] != 0 and fila[0] < 0.765:
                    parecida[num_columna] += 1
                num_columna += 1

            num_columna = 1
            for valor_dep in parecida:
                if valor_dep == 2:
                    noticias_parecidas.add(list_posiciones[num_matriz][num_columna])
                if valor_dep == 1:
                    distancia_columna = matriz[0][num_columna]
                    distancia_fila = matriz[num_columna][0]
                    if distancia_fila < 0.81 and distancia_columna < 0.81:
                        noticias_parecidas.add(list_posiciones[num_matriz][num_columna])
                if (matriz[0][num_columna] + matriz[num_columna][0])/2 < 0.781:
                    noticias_parecidas.add(list_posiciones[num_matriz][num_columna])
                num_columna += 1

            num_matriz += 1
        
        # Compruebo que no se trata del mismo articulo y que las fechas son relativamente del mismo tiempo
        noticias_a_eliminar = set()
        
        for noticia_parecida in noticias_parecidas:
            #print("Comprobando si la noticia " + nombre_noticia_buscada + " es la misma que la noticia " + noticia)
            f.write("La noticia es similar a la noticia "  + noticia_parecida + " teniendo en cuenta la NCD. Pero hay que realizar las siguientes comprobaciones:" + '\n')
            f.write(" - Comprobando si la noticia " + noticia + " es el mismo artículo que la noticia " + noticia_parecida + '\n')

            periodico_buscado = noticia.split("-")[0]
            noticia_buscada = noticia.split("-")[1]
            
            periodico1 = noticia_parecida.split("-")[0]
            noticia1 = noticia_parecida.split("-")[1]

            flag = 0 # Si cumplen ambas condiciones la flag tendra el valor 2

            if dicc_periodico_noticias[periodico1][noticia1]["url"] != dicc_periodico_noticias[periodico_buscado][noticia_buscada]["url"]:
                flag += 1
                f.write("     Se trata de noticias distintas: \n")
                f.write("        - " + dicc_periodico_noticias[periodico1][noticia1]["url"]+ '\n')
                f.write("        - " + dicc_periodico_noticias[periodico_buscado][noticia_buscada]["url"] + '\n\n')
            else:
                mismo_articulo.append(noticia)
                f.write("     Se trata de la misma noticia: \n")
                f.write("        - " + dicc_periodico_noticias[periodico1][noticia1]["url"]+ '\n')
                f.write("        - " + dicc_periodico_noticias[periodico_buscado][noticia_buscada]["url"]+ '\n\n')

            f.write(" - Comprobando si las fechas son relativamente del mismo tiempo, máximo 7 días de diferencia" + '\n')
            #print("Comprobando si las fechas son relativamente del mismo tiempo")
            if dicc_periodico_noticias[periodico1][noticia1]["fecha"] != "Muy antigua" and dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"] != "Muy antigua":
                fecha = dicc_periodico_noticias[periodico1][noticia1]["fecha"].split("-")[0] + "/" +dicc_periodico_noticias[periodico1][noticia1]["fecha"].split("-")[1] + "/" +dicc_periodico_noticias[periodico1][noticia1]["fecha"].split("-")[2].split("T")[0]
                fecha_buscada = dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"].split("-")[0] + "/" +dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"].split("-")[1] + "/" +dicc_periodico_noticias[periodico_buscado][noticia_buscada]["fecha"].split("-")[2].split("T")[0]
                fecha_1 = datetime.strptime(fecha, '%Y/%m/%d')
                fecha_2 = datetime.strptime(fecha_buscada, '%Y/%m/%d')

                if str(fecha_1 - fecha_2) == "0:00:00":
                    flag += 1
                    f.write("     Se trata de la misma fecha: \n")
                    f.write("        - " + fecha + '\n')
                    f.write("        - " + fecha_buscada + '\n\n')
                else:
                    #print(" Se diferencian en " + str(fecha_1 - fecha_2).split(" ")[0].split("-")[1] + " días")
                    if fecha_1 < fecha_2:
                        if int(str(fecha_1 - fecha_2).split(" ")[0].split("-")[1]) < 8:
                            flag += 1
                            #print("Son relativamente de la misma fecha.")
                    else:
                        if int(str(fecha_1 - fecha_2).split(" ")[0]) < 8:
                            flag += 1
                            #print("Son relativamente de la misma fecha.")
                    if fecha_1 < fecha_2:
                        f.write("     Tienen " + str(fecha_1 - fecha_2).split(" ")[0].split("-")[1] + " días de diferencia: \n")
                    else:
                        f.write("     Tienen " + str(fecha_1 - fecha_2).split(" ")[0] + " días de diferencia: \n")
                    f.write("        - " + fecha + '\n')
                    f.write("        - " + fecha_buscada + '\n\n')

            if flag != 2:
                noticias_a_eliminar.add(noticia_parecida)
                f.write("Demasiados días de diferencia o se trata del mismo artículo.\n\n")
            else:
                f.write("Cumple todos los aspectos, sí se trata de una noticia similar.\n\n")
                
            # Compruebo si se trata del mismo articulo que uno ya existente
            f.write("Comprobando si el mismo artículo ya se ha tenido en cuenta...\n")
            if noticia_parecida in mismo_articulo:
                noticias_a_eliminar.add(noticia_parecida)
                f.write("Este articulo ya se ha tenido en cuenta.\n\n")

        # Elimino las noticias que no cumplan las condiciones
        for n in noticias_a_eliminar:
            noticias_parecidas.remove(n)
            
        for n in noticias_parecidas:
            dicc_noticias_similares[noticia].add(n)
            
        f.write("Hay " + str(len(noticias_parecidas)) + " noticias similares:\n")
        f.write(str(noticias_parecidas) + "\n\n")
        
        f.write("Hasta el momento, hemos obtenido las siguientes dependencias:\n")
        f.write(str(dicc_noticias_similares) + "\n\n")

    f.close()
    return noticias_parecidas
```
